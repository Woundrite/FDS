```python
import pandas as pd
import os

files = [
    ('data/win7_normal_1.csv', 'win7'),
    ('data/win7_normal_2.csv', 'win7'),
    ('data/win7_normal_3.csv', 'win7'),
    ('data/win10_normal_1.csv', 'win10'),
    ('data/win10_normal_2.csv', 'win10'),
    ('data/win10_normal_3.csv', 'win10'),
    ('data/win10_normal_4.csv', 'win10'),
]

def clean_col(name):
    parts = name.strip().split('\\')
    parts = [p for p in parts if p]
    if len(parts) >= 2:
        return parts[-2].replace('(_Total)', '').strip() + '_' + parts[-1]
    return name

all_dfs = []

for filepath, os_label in files:
    print(f"Processing: {filepath}")

    df = pd.read_csv(filepath, low_memory=False)

    # Rename timestamp column
    df.rename(columns={df.columns[0]: 'timestamp'}, inplace=True)

    # Clean all column names
    df.columns = ['timestamp'] + [clean_col(c) for c in df.columns[1:]]

    # Remove blank first row
    df = df[df['timestamp'].str.strip() != ''].copy()

    # Parse timestamp
    df['timestamp'] = pd.to_datetime(df['timestamp'])

    # Convert all columns to numbers
    for col in df.columns:
        if col not in ['timestamp', 'os', 'label', 'source_file']:
            df[col] = pd.to_numeric(df[col], errors='coerce')

    # Add labels
    df['os'] = os_label
    df['label'] = 'normal'
    df['source_file'] = os.path.basename(filepath)

    all_dfs.append(df)
    print(f"  Done — {len(df)} rows, {df.shape[1]} columns")
```

```python
# Step 1 — Combine all files (keep only common columns)
combined = pd.concat(all_dfs, join='inner', ignore_index=True)
print(f"Combined shape: {combined.shape}")

# Step 2 — Fill missing values with column median
numeric_cols = combined.select_dtypes(include='number').columns
combined[numeric_cols] = combined[numeric_cols].fillna(combined[numeric_cols].median())
print(f"Missing values remaining: {combined.isnull().sum().sum()}")

# Step 3 — Remove constant/useless columns
constant_cols = [c for c in combined.columns if combined[c].nunique() <= 1]
combined.drop(columns=constant_cols, inplace=True)
print(f"Removed {len(constant_cols)} constant columns")

# Step 4 — Show final result
print(f"\nFinal dataset shape: {combined.shape}")
print(combined.head())
```

```python
# Save cleaned file back to your Google Drive
combined.to_csv('data/cleaned_combined.csv', index=False)

print("File saved successfully to your Google Drive!")
print(f"Location: data/cleaned_combined.csv")
print(f"Total rows: {len(combined)}")
print(f"Total columns: {combined.shape[1]}")
print(f"\nColumn names:")
for col in combined.columns:
    print(f"  - {col}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA, TruncatedSVD

file_path = 'data/cleaned_combined.csv'
df = pd.read_csv(file_path)
print("Dataset loaded successfully!")
print("Shape:", df.shape)
print(df.head())

# ============================================================
# 2. PREPARE DATA FOR REDUCTION
# ============================================================
# Convert timestamp if present
if 'timestamp' in df.columns:
    df['timestamp'] = pd.to_datetime(df['timestamp'], errors='coerce')

# Keep only numeric columns for PCA/SVD
numeric_df = df.select_dtypes(include=[np.number]).copy()

print("\nNumeric feature shape before reduction:", numeric_df.shape)
print("Numeric columns:")
print(list(numeric_df.columns))

# Optional: Remove any remaining constant columns (safety check)
constant_cols = [col for col in numeric_df.columns if numeric_df[col].nunique() <= 1]
if constant_cols:
    numeric_df.drop(columns=constant_cols, inplace=True)
    print(f"\n Removed {len(constant_cols)} remaining constant columns")
else:
    print("\n No remaining constant numeric columns")

print("Final numeric feature shape:", numeric_df.shape)

# ============================================================
# 3. HANDLE MISSING VALUES (safety check)
# ============================================================
numeric_df = numeric_df.fillna(numeric_df.median())

# ============================================================
# 4. STANDARDIZE FEATURES (VERY IMPORTANT FOR PCA & SVD)
# ============================================================
scaler = StandardScaler()
X_scaled = scaler.fit_transform(numeric_df)

print("\n Data standardized successfully!")
print("Scaled data shape:", X_scaled.shape)

# ============================================================
# 5. PCA - FIND COMPONENTS FOR 95% VARIANCE
# ============================================================
pca_full = PCA()
pca_full.fit(X_scaled)

explained_variance_ratio = pca_full.explained_variance_ratio_
cumulative_variance = np.cumsum(explained_variance_ratio)

# Find number of components for 95% variance
pca_k = np.argmax(cumulative_variance >= 0.95) + 1

print("\n================ PCA RESULTS ================")
print(f"Total original numeric features: {X_scaled.shape[1]}")
print(f"Number of PCA components for 95% variance: {pca_k}")
print(f"Cumulative variance retained: {cumulative_variance[pca_k - 1]:.4f}")

# Fit final PCA
pca = PCA(n_components=pca_k)
X_pca = pca.fit_transform(X_scaled)

print("Reduced PCA shape:", X_pca.shape)

# Create PCA dataframe
pca_columns = [f'PC{i+1}' for i in range(X_pca.shape[1])]
pca_df = pd.DataFrame(X_pca, columns=pca_columns)

# Add optional metadata columns back for reference
for col in ['timestamp', 'os', 'source_file']:
    if col in df.columns:
        pca_df[col] = df[col].values

# Save PCA reduced dataset
pca_output_path = 'data/pca_reduced_95.csv'
pca_df.to_csv(pca_output_path, index=False)

print(f"PCA reduced dataset saved at: {pca_output_path}")

# ============================================================
# 6. TRUNCATED SVD - FIND COMPONENTS FOR 95% VARIANCE
# ============================================================
# TruncatedSVD works well on numeric matrices and can also be used for reduction
# We test components from 1 to total_features to find 95% variance

n_features = X_scaled.shape[1]
svd_variance_list = []

for k in range(1, n_features + 1):
    svd_temp = TruncatedSVD(n_components=k, random_state=42)
    svd_temp.fit(X_scaled)
    svd_variance_list.append(svd_temp.explained_variance_ratio_.sum())

svd_cumulative_variance = np.array(svd_variance_list)

# Find number of SVD components for 95% variance
svd_k = np.argmax(svd_cumulative_variance >= 0.95) + 1

print("\n================ SVD RESULTS ================")
print(f"Total original numeric features: {X_scaled.shape[1]}")
print(f"Number of SVD components for 95% variance: {svd_k}")
print(f"Cumulative variance retained: {svd_cumulative_variance[svd_k - 1]:.4f}")

# Fit final TruncatedSVD
svd = TruncatedSVD(n_components=svd_k, random_state=42)
X_svd = svd.fit_transform(X_scaled)

print("Reduced SVD shape:", X_svd.shape)

# Create SVD dataframe
svd_columns = [f'SVD{i+1}' for i in range(X_svd.shape[1])]
svd_df = pd.DataFrame(X_svd, columns=svd_columns)

# Add optional metadata columns back for reference
for col in ['timestamp', 'os', 'source_file']:
    if col in df.columns:
        svd_df[col] = df[col].values

# Save SVD reduced dataset
svd_output_path = 'data/svd_reduced_95.csv'
svd_df.to_csv(svd_output_path, index=False)

print(f"SVD reduced dataset saved at: {svd_output_path}")

# ============================================================
# 7. COMPARISON SUMMARY
# ============================================================
comparison_df = pd.DataFrame({
    'Method': ['Original', 'PCA (95%)', 'SVD (95%)'],
    'Number_of_Features': [X_scaled.shape[1], pca_k, svd_k],
    'Variance_Retained': [1.0, cumulative_variance[pca_k - 1], svd_cumulative_variance[svd_k - 1]]
})

print("\n================ COMPARISON SUMMARY ================")
print(comparison_df)

comparison_output_path = 'data/reduction_comparison_summary.csv'
comparison_df.to_csv(comparison_output_path, index=False)

print(f"Comparison summary saved at: {comparison_output_path}")

# ============================================================
# 8. PLOT - PCA EXPLAINED VARIANCE
# ============================================================
plt.figure(figsize=(10, 5))
plt.plot(range(1, len(cumulative_variance) + 1), cumulative_variance, marker='o')
plt.axhline(y=0.95, color='r', linestyle='--', label='95% Variance Threshold')
plt.axvline(x=pca_k, color='g', linestyle='--', label=f'PCA Components = {pca_k}')
plt.title('PCA Cumulative Explained Variance')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.legend()
plt.grid(True)
plt.show()

# ============================================================
# 9. PLOT - SVD EXPLAINED VARIANCE
# ============================================================
plt.figure(figsize=(10, 5))
plt.plot(range(1, len(svd_cumulative_variance) + 1), svd_cumulative_variance, marker='o')
plt.axhline(y=0.95, color='r', linestyle='--', label='95% Variance Threshold')
plt.axvline(x=svd_k, color='g', linestyle='--', label=f'SVD Components = {svd_k}')
plt.title('SVD Cumulative Explained Variance')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.legend()
plt.grid(True)
plt.show()

# ============================================================
# 10. OPTIONAL - SAVE LOADINGS / FEATURE IMPORTANCE FOR PCA
# ============================================================
# PCA loadings show contribution of original features to each principal component
loadings = pd.DataFrame(
    pca.components_.T,
    columns=[f'PC{i+1}' for i in range(pca_k)],
    index=numeric_df.columns
)

loadings_output_path = 'data/pca_loadings.csv'
loadings.to_csv(loadings_output_path)

print(f"\n PCA loadings saved at: {loadings_output_path}")

# ============================================================
# 11. FINAL MESSAGE
# ============================================================
print("\n DATA REDUCTION COMPLETED SUCCESSFULLY!")
print(f"Original numeric features: {X_scaled.shape[1]}")
print(f"PCA reduced features: {pca_k}")
print(f"SVD reduced features: {svd_k}")
print("\nSaved files:")
print(" - PCA reduced dataset: pca_reduced_95.csv")
print(" - SVD reduced dataset: svd_reduced_95.csv")
print(" - Comparison summary: reduction_comparison_summary.csv")
print(" - PCA loadings: pca_loadings.csv")
```
